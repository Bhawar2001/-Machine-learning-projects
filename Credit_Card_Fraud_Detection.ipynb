Importing The Dependenciesm


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

credit_card_creditcard = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_creditcard['Class'].value_counts()
```

This Dataset is highly unblanced

0 --> Normal Transaction

1 --> fraudulent transaction

```python
legit = credit_card_creditcard[credit_card_creditcard.Class == 0]
fraud = credit_card_creditcard[credit_card_creditcard.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_creditcard.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_creditcard = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_creditcard.head()
```

```python
new_creditcard.tail()
```

```python
new_creditcard.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_creditcard.drop(columns='Class', axis=1)
Y = new_creditcard['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
